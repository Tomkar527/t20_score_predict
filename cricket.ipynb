```python
import pandas as pd
import numpy as np
```

```python
df = df.drop(columns=['Unnamed: 0','tossWinner'])
```

```python
df['battingTeam'].value_counts()
```

```python
df = df.drop(columns=['city','venue'])
```

```python
df.head()
```

```python
df['matchId']=df['matchId'].str.replace('_1','').astype('int')
```

```python
df
```

```python
df.info()
```

```python
df['current_score'] = df.groupby('matchId')['runs'].cumsum()
```

```python
df
```

```python
df['overs'] = df['balls'].astype(int)
df['ball_number'] = round((df['balls'] % 1) * 10).astype(int)
```

```python
df
```

```python
df['ball_bowled'] = (df['overs'].astype('int')*6) + df['ball_number'].astype('int')
```

```python
df
```

```python
df['crr']= (df['current_score']*6)/df['ball_bowled']
```

```python
df
```

```python
df
```

```python
final_df = df.groupby('matchId').sum()['runs'].reset_index().merge(df,on='matchId')
```

```python
final_df
```

```python
final_df['wickets'] = final_df.groupby('matchId')['player_out'].cumsum()
```

```python
final_df
```

```python
# batting team
# bowling team
# current score
# wickets
# over
# runx
```

```python
final = final_df[['battingTeam','bowlingTeam','current_score','balls','crr','wickets','runs_x']]
```

```python
final
```

```python
final['Total_runs'] = final['runs_x']
final = final.drop(columns = ['runs_x'])
```

```python
final = final.sample(frac=1).reset_index(drop=True)
```

```python
final.head()
```

```python
final.to_csv("Cricket_score_predictor.csv",index=False)
```
